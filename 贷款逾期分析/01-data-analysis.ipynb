```python
import pandas as pd
import numpy as np
```

这份数据集是金融数据（非原始数据，已经处理过了），我们要做的是预测贷款用户是否会逾期。表格中 "status" 是结果标签：0表示未逾期，1表示逾期。

```python
data = pd.read_csv("./../data.csv", encoding='gbk')
# utf-8编码报错了，说是不支持，'utf-8' codec can't decode byte 0xbf in position 0: invalid start byte
```

```python
data.head()
```

bank_card_no那一列很可能全是“卡号1”，明显是人为处理过来保密用户信息的，对最终结果预测也没有影响，属于无关特征，可以删去

```python
def deleteIrrelevantFeature(data): # 删去没用的列
    for col in data.columns:
        if len(data[col].value_counts()) == 1: # 如果这一列的值全都一样，那它就没用
            print(col)
            data.drop(columns=[col], inplace=True)
```

```python
deleteIrrelevantFeature(data)
data.head()
```

经过人为观察发现，“Unnamed: 0", "custid", "trade_no", "id_name"这几列也都没用，删去

```python
data.drop(columns=["Unnamed: 0", "custid", "trade_no", "id_name"], inplace=True)
data.head()
```

```python
data.info()
```

```python
# 查看剩下的obiect类型有哪些
for i in data:
    if data[i].dtype == 'object':
        print(i)
```

```python
# 时间类型的转换
data["latest_query_time"] = pd.to_datetime(data["latest_query_time"])
data["loans_latest_time"] = pd.to_datetime(data["loans_latest_time"])
data["first_transaction_time"] = pd.to_datetime(data["first_transaction_time"]) #被误判成float
```

```python
# data.reg_preference_for_trad.unique()
data['reg_preference_for_trad'].value_counts(dropna = False)
```

```python
# 把Nan当作其它类型，然后采取对其编码
data['reg_preference_for_trad'].fillna('其他城市',inplace = True)
data['reg_preference_for_trad'].replace({'一线城市':1,'二线城市':2,'三线城市':3,'境外':4,'其他城市':5},inplace = True)
```

至此，已经完成了无关特征的删除和数据格式的转换，下面进行缺失值处理

```python
data.isnull().sum(axis=0).sort_values(ascending=False) # 查看每一列缺失的数据德尔个数，发现student_feature缺的异常多
```

```python
print(data["student_feature"][0:100]) # 单独查看student_feature这一列，发现要么是1.0要么是NaN，推测NaN就是不是学生，用0替换
data["student_feature"] = data["student_feature"].fillna(0)
```

```python
# 其余的缺失值太多的话直接进行删除
data.dropna(axis=0,thresh = data.shape[1]*0.7, inplace=True)
data.dropna(axis=1,thresh = data.shape[0]*0.7, inplace=True)
# 然后用均值填充缺失值
data.fillna(data.mean(), inplace=True)
```
